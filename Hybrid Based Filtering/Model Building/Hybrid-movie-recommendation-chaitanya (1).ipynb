```python
col=np.array(links['tmdbId'], np.int64)
links['tmdbId']=col
```

```python
hoo.rename(columns={'id':'tmdbId'}, inplace=True)
hoo = pd.merge(hoo,links,on='tmdbId')
hoo.drop(['imdb_id'], axis=1, inplace=True)
hoo.head()
```

```python
tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(hoo['overview'])
tfidf_matrix.shape
```

```python

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

indices = pd.Series(hoo.index, index=hoo['original_title']).drop_duplicates()
```

```python
def recommend(title, cosine_sim=cosine_sim):
    
    
    idx = indices[title]
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    sim_scores = sim_scores[1:16]
    
    movie_indices = [i[0] for i in sim_scores]
    
    for i in movie_indices:
        pop = hoo.at[i,'vote_average']
        if pop<5 or pop>10:
            movie_indices.remove(i)

    
    return hoo[['original_title','vote_average']].iloc[movie_indices]
```

```python
recommend('Iron Man')
```

```python
reader = Reader() 
df = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
kf = KFold(n_splits=5)
kf.split(df) 
```

```python
svd = SVD()
cross_validate(svd, df, measures=['RMSE', 'MAE'])

trainset = df.build_full_trainset()
svd.fit(trainset)
```

```python
links_df = pd.read_csv('/home/chaitanyadubal/Downloads/links_small.csv')
col=np.array(links_df['tmdbId'], np.int64)
links_df['tmdbId']=col


links_df = links_df.merge(hoo[['title', 'tmdbId']], on='tmdbId').set_index('title')
links_index = links_df.set_index('tmdbId') 
```

```python
def hybrid(userId, title):
    idx = indices[title]
    tmdbId = links_df.loc[title]['tmdbId']
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:31] 
    movie_indices = [i[0] for i in sim_scores]
    
    movies = hoo.iloc[movie_indices][['title', 'vote_average', 'tmdbId']]
    movies['est'] = movies['tmdbId'].apply(lambda x: svd.predict(userId, links_index.loc[x]['movieId']).est)
    movies = movies.sort_values('est', ascending=False) 
    movies.columns = ['Title', 'Vote Average', 'TMDb Id', 'Estimated Prediction']
    return movies.head(15)
```

```python
hybrid(1, 'The Conjuring')
```
